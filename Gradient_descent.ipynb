```python
import graphlab as gl
```

```python
import numpy as np
sales=gl.SFrame('kc_house_data.gl')
```

```python
def numpy_data(sales , input_features ,output):
    sales['constant']=1
    input_features=['constant']+input_features
    features_sales=sales.select_columns(input_features)
    feature_matrix=features_sales.to_numpy()
    target_sarray=sales[output]
    target_numarray=target_sarray.to_numpy()
    
    return(feature_matrix , target_numarray)

```

```python
def predict_output(feature_matrix , weights):
    prediction=np.dot(feature_matrix,weights)
    
    return(prediction)
```

```python
def feature_derivative(errors, feature):
    
    derivative=np.dot(errors,feature)
    derivative=2*derivative
    
    return(derivative)

(example_features, example_output) = numpy_data(sales, ['sqft_living'], 'price') 
my_weights = np.array([0., 0.])
test_predictions = predict_output(example_features, my_weights) 
errors=test_predictions-example_output
feature=example_features[:,0]
derivate=feature_derivative(errors, feature)
print -np.sum(example_output)*2
derivate
```

```python
from math import sqrt
def regression_gradient_descent(feature_matrix, output, initial_weights, step_size, tolerance):
    
    converged=False
    weights=np.array(initial_weights)
 
   
    while not converged:
        prediction=predict_output(feature_matrix , weights)
        errors=prediction-output
        
        gradient_sum_squares = 0
        
        for i in range(len(weights)):
            feature_col=feature_matrix[:,i]
            derivate=feature_derivative(errors ,feature_col)
            gradient_sum_squares+=(derivate*derivate)
            weights[i]-=step_size*derivate
            
        gradient_magnitude=sqrt(gradient_sum_squares)
        print " gradient magnitude :"
        print gradient_magnitude
        if gradient_magnitude<tolerance:
            converged=True
        
        
    return(weights)
        
    
```

```python
train_data,test_data = sales.random_split(.8,seed=0)
```

```python
simple_features = ['sqft_living']
my_output = 'price'
(simple_feature_matrix, output) = numpy_data(train_data, simple_features, my_output)
initial_weights = np.array([-47000., 1.])
step_size = 7e-12
tolerance = 2.5e7
simple_feature_matrix ,output
```

```python
simple_weight=regression_gradient_descent(simple_feature_matrix,output,[-47000., 1.],step_size,tolerance)
```

```python
(test_simple_feature_matrix, test_output) =numpy_data(test_data, simple_features, my_output)
```

```python
test_prediction=np.dot(test_simple_feature_matrix,simple_weight)
```

```python
model_features = ['sqft_living', 'sqft_living15']
output = 'price'
(final_features_matrix ,final_output)=numpy_data(train_data,model_features ,output)
initial_weights = np.array([-100000., 1., 1.])
step_size = 4e-12
tolerance = 1e9
```
